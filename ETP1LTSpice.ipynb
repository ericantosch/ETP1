```python
from matplotlib import pyplot as plt
import math
import numpy as np
import pandas as pd
LT = pd.read_excel('LTSpiceData.xlsx')
```

```python
plt.rc('text', usetex=True)
plt.rc('font', family='serif')
fig = plt.figure()
ax1 = fig.add_subplot(111)
ax1.set_ylabel('Strom in A')
LT.plot(x='Voltage', y='Current', color='g', ax=ax1, label='_Power')
ax1.set_xlabel('Spannung in V')
plt.title(r'\textbf{Die Spannungsquellenkennlinie}', fontsize=11)
plt.xlim(0, 7)
plt.ylim(0, 0.0005)
ax1.yaxis.label.set_color('g')
plt.savefig('curr_volt.png')

```

```python
from matplotlib import pyplot as plt
import math
import numpy as np
import pandas as pd
LT = pd.read_excel('LTSpiceData2.xlsx')
plt.rc('text', usetex=True)
plt.rc('font', family='serif')
fig = plt.figure()
host = fig.add_subplot(111)

par1 = host.twinx()
par2 = host.twinx()

host.set_xlabel(r'\textbf{Widerstand in $\Omega$}', fontsize=11)
host.set_ylabel("Power in W")
par1.set_ylabel("Voltage in V")
par2.set_ylabel("Current in A")


sp1 = LT.head(30).plot(x='Resistance', y='Power', color='g', ax=host, label = '_Power', legend = None)
sp2 = LT.head(30).plot(x='Resistance', y='Voltage', color='r', ax=par1, label = '_Voltage', legend = None)
sp3 = LT.head(30).plot(x='Resistance', y='Current', color='b', ax=par2, label = '_Current', legend = None)
par2.spines['right'].set_position(('outward', 60)) 
lns = [sp1, sp2, sp3]
host.yaxis.label.set_color('g')
par1.yaxis.label.set_color('r')
par2.yaxis.label.set_color('b')

plt.title(r'\textbf{Die Leistung und Spannung der Last}', fontsize=11)

plt.savefig('pw_volt_rsjp2.pdf')
```

```python
plt.rc('text', usetex=True)
plt.rc('font', family='serif')
fig = plt.figure()
ax1 = fig.add_subplot(111)
ax1.set_ylabel('Leistung in W')
ax2 = ax1.twinx()
ax2.set_ylabel('Spannung in V')
LT.head(30).plot(x='Resistance', y='Power', color='g', ax=ax1, label='_Power')
LT.head(30).plot(x='Resistance', y='Voltage', color='r', ax=ax2, label='_Voltage')
plt.title(r'\textbf{Die Leistung und Spannung der Last}', fontsize=11)
ax1.yaxis.label.set_color('g')
ax2.yaxis.label.set_color('r')
plt.savefig('pw_volt2.png')

```
